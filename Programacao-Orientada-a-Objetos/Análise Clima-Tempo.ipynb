```python
# biblioteca para pegar uma página HTML
import requests
```

```python
# Manipula a página HTML
from bs4 import BeautifulSoup
```

```python
# Faz a limpeza e mostra as informações
import pandas as pd
```

```python
# Faz a manipulação numérica
import numpy as np
```

```python
url ="https://forecast.weather.gov/MapClick.php?lat=42.93708397900008&lon=-75.61070144699994"
```

```python
# faz o dowloand da pagina, se a resposta for 200 está correto
page = requests.get( url )
page
```

```python
#page.content lê o conteúdo do arqueivo o o html.parser é o tipo de analisador
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
#find busca o conteúdo da palavra entre parênteses
seven_day = soup.find(id="seven-day-forecast")
print(seven_day.prettify())
```

```python
#fin_all é para pegar todas as tombstoone-container
forecast_= seven_day.find_all(class_="tombstone-container")
```

```python
print(forecast_)
```

```python
morning=forecast_[0]
morning
```

```python
print(morning.prettify())
```

```python
period = morning.find(class_="period-name").get_text()
period
```

```python
short_desc=morning.find(class_="short-desc").get_text()
short_desc
```

```python
temp = morning.find(class_="temp").get_text()
temp
```

```python
seven_day
```

```python
# mesmo que find mas mais específico
seven_day.select(".tombstone-container .period-name")
```

```python
period_tags = seven_day.select(".tombstone-container .period-name")
# o get_text() é para pegar só o texto contido no html
periods = []
for pt in period_tags:
    periods.append(pt.get_text())
periods
```

```python
short_descs_tags = seven_day.select(".tombstone-container .short-desc")
short_descs = []
for sd in short_descs_tags:
    short_descs.append(sd.get_text())
short_descs
```

```python
temps_tags = seven_day.select(".tombstone-container .temp")
temps = [] 
for tp in temps_tags:
    temps.append(tp.get_text())
temps
```

```python
descs_tags = seven_day.select(".tombstone-container img")
descs = []
for d in descs_tags:
    descs.append(d['title'])
descs
```

```python
df_clima_tempo = pd.DataFrame({
    "Período": periods,
    "Minidescrição": short_descs,
    "Temperatura": temps,
    "Descrição":descs
})
df_clima_tempo
```

```python
df_clima_tempo["Temperatura"]
```

```python
temp_nums = df_clima_tempo["Temperatura"].str.extract(r"(\d+)", expand=False)
temp_nums
```

```python
df_clima_tempo["Temperatura"] = temp_nums.astype('int')
df_clima_tempo["Temperatura"]
```

```python
#Calcula a média para as temperaturas altas e baixas.
df_clima_tempo["Temperatura"].mean()
```
